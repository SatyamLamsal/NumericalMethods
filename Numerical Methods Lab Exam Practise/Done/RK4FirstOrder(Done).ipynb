```python
def F(x,y):
    return x**2 - y
```

```python
import numpy as np
import matplotlib.pyplot as plt
def RK4(x0, y0, xn, n):
    h = (xn-x0) / n
    X = np.zeros(n+1, dtype=float)
    Y = np.zeros(n+1, dtype=float)
    X[0], Y[0] = x0, y0
    for i in range(n):
        k1 = h * F(X[i], Y[i])
        k2 = h * F(X[i] + h/2, Y[i] + k1/2)
        k3 = h * F(X[i] + h/2, Y[i] + k2/2)
        k4 = h * F(X[i] + h, Y[i] + k3)
        k = (k1 + 2*k2 + 2*k3 + k4)/6
        X[i+1] = X[i] + h
        Y[i+1] = Y[i] + k
    return X, Y
```

```python

x0, y0, xn, n = -10, 122, 12, 100
X, Y = RK4(x0, y0, xn, n)
for i in range(n+1):
    print(f"{X[i]:12.4f} {Y[i]:12.4f}")
```

```python
plt.title("RK4 Method in action")
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```
